```python
# Define o caminho base para os dados raw
RAW_PATH = '../data/raw/'

# Carrega todas as tabelas
orders = pd.read_csv(RAW_PATH + 'olist_orders_dataset.csv')
order_items = pd.read_csv(RAW_PATH + 'olist_order_items_dataset.csv')
customers = pd.read_csv(RAW_PATH + 'olist_customers_dataset.csv')
products = pd.read_csv(RAW_PATH + 'olist_products_dataset.csv')
sellers = pd.read_csv(RAW_PATH + 'olist_sellers_dataset.csv')
reviews = pd.read_csv(RAW_PATH + 'olist_order_reviews_dataset.csv')
payments = pd.read_csv(RAW_PATH + 'olist_order_payments_dataset.csv')
geolocation = pd.read_csv(RAW_PATH + 'olist_geolocation_dataset.csv')
product_category_translation = pd.read_csv(RAW_PATH + 'product_category_name_translation.csv')

print("Todas as tabelas carregadas com sucesso.")
```

```python
def inventario_tabela(df, nome):
    """
    Gera um inventário completo de uma tabela.
    Sempre executar isso antes de qualquer transformação.
    """
    print(f"\n{'='*60}")
    print(f"TABELA: {nome.upper()}")
    print(f"{'='*60}")
    print(f"Shape: {df.shape[0]:,} linhas × {df.shape[1]} colunas")
    print(f"\n--- Tipos e nulos ---")
    
    info_df = pd.DataFrame({
        'coluna': df.columns,
        'tipo': df.dtypes.values,
        'nulos': df.isnull().sum().values,
        'pct_nulo': (df.isnull().sum().values / len(df) * 100).round(2),
        'únicos': df.nunique().values,
        'exemplo': [df[col].dropna().iloc[0] if df[col].notna().any() else 'NaN' for col in df.columns]
    })
    
    print(info_df.to_string(index=False))
    print(f"\n--- Duplicatas ---")
    print(f"Linhas duplicadas: {df.duplicated().sum():,}")
    
    return info_df

# Executando a função para cada uma das tabelas carregadas
inv_orders = inventario_tabela(orders, 'orders')
inv_items = inventario_tabela(order_items, 'order_items')
inv_customers = inventario_tabela(customers, 'customers')
inv_products = inventario_tabela(products, 'products')
inv_sellers = inventario_tabela(sellers, 'sellers')
inv_reviews = inventario_tabela(reviews, 'reviews')
inv_payments = inventario_tabela(payments, 'payments')
```
